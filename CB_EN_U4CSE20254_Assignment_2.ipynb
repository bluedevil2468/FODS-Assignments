```python
import random as rd
import time as t
import matplotlib.pyplot as plt

def times(n):
    l=[]
    for i in range(n):
        l.append(rd.randint(0,1000))
    ti = selectionsort(l)
    return (ti)

def timebs(n):
    l=[]
    for i in range(n):
        l.append(rd.randint(0,1000))
    ti = bubblesort(l)
    return (ti)

def timeis(n):
    l=[]
    for i in range(n):
        l.append(rd.randint(0,1000))
    ti = insertionsort(l)
    return (ti)

def timems(n):
    l=[]
    for i in range(n):
        l.append(rd.randint(0,1000))
    ti = mergesort(l)
    return (ti)

def timeqs(n):
  l=[]
  for i in range(n):
        l.append(rd.randint(0,1000))
  start = t.time()
  ti = quick(l,0,len(l)-1)
  end = t.time()
  return (end-start)


def selectionsort(l):
    start = t.time()
    for i in range(len(l)):
        mini = i
        for j in range(i,len(l)):
            if (l[mini]>l[j]):
                mini = j
        l[i],l[mini] = l[mini],l[i]
    end = t.time()
    return (end-start)

def bubblesort(l):
  start = t.time()
  for i in range(len(l)):
      for j in range(len(l)-i-1):
          if (l[j]>l[j+1]):
              l[j],l[j+1] = l[j+1],l[j]
  end = t.time()
  return (end-start)

def insertionsort(l):
  start = t.time()
  for i in range(len(l)):
      el = l[i]
      j = i-1
      while j >= 0 and el < l[j] :
          l[j + 1] = l[j]
          j -= 1
          l[j + 1] = el
  end = t.time()
  return (end-start)

def mergesort(l):
  start = t.time()
  if(len(l)>1):
    mid=len(l)//2
    L=l[:mid]
    R=l[mid:]
    mergesort(L)
    mergesort(R)
    i=0
    j=0
    k=0
    while(i<len(L) and j<len(R)):
      if(L[i]<R[j]):
        l[k]=L[i]
        i+=1
      else:
        l[k]=R[j]
        j+=1
      k+=1
    while(i<len(L)):
      l[k]=L[i]
      i+=1
      k+=1
    while(j<len(R)):
      l[k]=R[j]
      j+=1
      k+=1
  end = t.time()
  return (end-start)

def partition(l,st,en):
    i=(st-1)
    pivot=l[en]
    for j in range(st,en):
        if(l[j]<=pivot):
            i+=1
            l[i],l[j]=l[j],l[i]
            
    l[i+1],l[en]=l[en],l[i+1]
    return i+1

def quick(l,st,en):
    if(st<en):
        p=partition(l,st,en)
        quick(l,st,p-1)
        quick(l,p+1,en)
  
x = []
y = []
a = []
b=[]
c=[]
d=[]
e=[]
f=[]
g=[]
h=[]
q=10
while(q<=10000):
    x.append(q)
    z = times(q)
    y.append(z)
    q+=100
q = 10
while(q<=10000):
    a.append(q)
    z = timebs(q)
    b.append(z)
    q+=100
q = 10
while(q<=10000):
    c.append(q)
    z = timeis(q)
    d.append(z)
    q+=100
q = 10
while(q<=10000):
    e.append(q)
    z = timeqs(q)
    f.append(z)
    q+=100
q = 10
while(q<=10000):
    g.append(q)
    z = timems(q)
    h.append(z)
    q+=100

plt.ylabel("Time")
plt.xlabel("Input size")
plt.plot(x,y,color='k',label='Selection Sort') #selection
plt.plot(a,b,color='blue',label='Bubble Sort') #bubble
plt.plot(c,d,color='green',label='Insertion Sort') #insertion
plt.plot(e,f,color='orange',label='Quick Sort') #quick
plt.plot(g,h,color='red',label='Merge Sort', ls = '--') #merge
plt.legend()
plt.show()



    
```
